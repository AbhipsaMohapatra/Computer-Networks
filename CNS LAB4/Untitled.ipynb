```python
key = input("Enter plain text").upper()
alpha=['A','B','C','D','E','F','G','H','I','K','L','M','N','O','P','Q','R','S','T','U','V','W','X','Y','Z']
for i in key:
    if(i in alpha):
        alpha.remove(i)

print(alpha)
t=0
k=0
mat=list()
for i in range(1,6):
    temp=list();
    while(t<len(key)):
        temp.append(key[t])
        t+=1
    while(k<len(alpha)):
        temp.append(alpha[k])
        
    mat.append(temp) 
    
print(mat)        
```
